```python
def lcm(x,y):
    if x>y:
        small=y
    else:
        small=x
        for i in range(small,x*y+1):
            if i%x==0 and i%y==0:
                return'LCM=',i
                break
lcm(14,21)            
```

```python
#Count the number of digits of an input integer by using while loop.
n=int(input('n=')) #102420
count=0
while n>0: #102420, 10242
    count=count+1
    n=n//10 #102420//10=10242, 10242//10=1024,1024//10=102, 102//=10,10//10=1,1//10=0
print(count)
```

```python
#Write a python program to add the digits of an input positive integer by using while loop.
n=int(input('n=')) #n=234, 2+3+4
remas=0
while n>0:
    x=n%10 #x=4,3,2
    remas=remas+x
    n=n//10 #234//10=23,23//10=2,2//10=0
print(remas)
```
